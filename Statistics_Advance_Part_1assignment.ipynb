# **STATISTICS ADVANCE PART 1 THEORY QUESTIONS**

# 1. WHAT IS A RANDOM VARIABLE IN PROBABILITY THEORY?

  - A RANDOM VARIABLE IS A VARIABLE WHOSE VALUE IS SUBJECT TO VARIATIONS DUE TO CHANCE. IT REPRESENTS THE OUTCOMES OF A RANDOM PHENOMENON. FOR EXAMPLE, IF YOU ROLL A DIE, THE NUMBER THAT COMES UP IS A RANDOM VARIABLE, BECAUSE IT’S UNCERTAIN UNTIL THE DIE IS ROLLED.

# 2. WHAT ARE THE TYPES OF RANDOM VARIABLES?

  • THERE ARE TWO MAIN TYPES OF RANDOM VARIABLES:

  • DISCRETE RANDOM VARIABLES — THESE TAKE ON A FINITE OR COUNTABLE NUMBER OF VALUES. AN EXAMPLE IS THE NUMBER OF HEADS WHEN FLIPPING A COIN MULTIPLE TIMES.

  • CONTINUOUS RANDOM VARIABLES — THESE TAKE ON AN INFINITE NUMBER OF VALUES WITHIN A GIVEN RANGE. AN EXAMPLEIS THE HEIGHT OF INDIVIDUALS IN A POPULATION, WHICH CAN VARY CONTINUOUSLY.

# 3. WHAT IS THE DIFFERENCE BETWEEN DISCRETE AND CONTINUOUS DISTRIBUTIONS?

  - DISCRETE DISTRIBUTIONS APPLY TO DISCRETE RANDOM VARIABLES WHERE THE DATA POINTS ARE COUNTABLE. EXAMPLES INCLUDE THE BINOMIAL DISTRIBUTION AND POISSON DISTRIBUTION. CONTINUOUS DISTRIBUTIONS, ON THE OTHER HAND, APPLY TO CONTINUOUS RANDOM VARIABLES, WHERE DATA POINTS CAN TAKE ON ANY VALUE WITHIN A RANGE. EXAMPLES INCLUDE THE NORMAL DISTRIBUTION AND THE UNIFORM DISTRIBUTION.

# 4. WHAT ARE PROBABILITY DISTRIBUTION FUNCTIONS (PDF)?

  - A PROBABILITY DISTRIBUTION FUNCTION (PDF) IS A FUNCTION THAT DESCRIBES THE LIKELIHOOD OF A RANDOM VARIABLE TAKING ON A PARTICULAR VALUE. FOR DISCRETE RANDOM VARIABLES, THIS FUNCTION GIVES THE PROBABILITY OF EACH POSSIBLE OUTCOME. FOR CONTINUOUS RANDOM VARIABLES, THE PDF GIVES THE LIKELIHOOD OF THE RANDOM VARIABLE FALLING WITHIN A SPECIFIC RANGE OF VALUES, BUT IT CANNOT GIVE THE PROBABILITY OF A SINGLE EXACT VALUE (AS IT’S 0 FOR ANY SPECIFIC POINT).

# 5. HOW DO CUMULATIVE DISTRIBUTION FUNCTIONS (CDF) DIFFER FROM PROBABILITY DISTRIBUTION FUNCTIONS (PDF)?

  - THE CDF IS RELATED TO THE PDF BUT PROVIDES CUMULATIVE PROBABILITY. WHILE THE PDF GIVES THE PROBABILITY OF A SPECIFIC OUTCOME, THE CDF GIVES THE PROBABILITY THAT A RANDOM VARIABLE IS LESS THAN OR EQUAL TO A PARTICULAR VALUE. IN OTHER WORDS, THE CDF SUMS THE PROBABILITIES UP TO THAT POINT.

# 6. WHAT IS A DISCRETE UNIFORM DISTRIBUTION?

  -  A DISCRETE UNIFORM DISTRIBUTION IS A TYPE OF DISTRIBUTION WHERE ALL OUTCOMES ARE EQUALLY LIKELY. FOR EXAMPLE, WHEN ROLLING A FAIR SIX-SIDED DIE, EACH FACE (1 THROUGH 6) HAS AN EQUAL PROBABILITY OF APPEARING, SO IT FOLLOWS A DISCRETE UNIFORM DISTRIBUTION.

# 7. WHAT ARE THE KEY PROPERTIES OF A BERNOULLI DISTRIBUTION?

  • A BERNOULLI DISTRIBUTION MODELS A RANDOM EXPERIMENT WITH EXACTLY TWO POSSIBLE OUTCOMES: SUCCESS (USUALLY CODED AS 1) AND FAILURE (CODED AS O). THE KEY PROPERTIES INCLUDE:

  • IT'S A DISCRETE DISTRIBUTION.

  • THE PROBABILITY OF SUCCESS IS DENOTED BY P, AND THE PROBABILITY OF FAILURE IS 1-P.

  • IT HAS A MEAN OF P AND A VARIANCE OF P(L-P).

# 8. WHAT IS THE BINOMIAL DISTRIBUTION, AND HOW IS IT USED IN PROBABILITY?

  -  THE BINOMIAL DISTRIBUTION DESCRIBES THE NUMBER OF SUCCESSES IN A FIXED NUMBER OF INDEPENDENT BERNOULLI TRIALS, EACH WITH THE SAME PROBABILITY OF SUCCESS. IT IS USED WHEN THERE ARE TWO POSSIBLE OUTCOMES (SUCCESS OR FAILURE) IN A FIXED NUMBER OF TRIALS, AND THE TRIALS ARE INDEPENDENT OF EACH OTHER. FOR EXAMPLE, IT CAN BE USED TO MODEL THE NUMBER OF HEADS IN 10 FLIPS OF A FAIR COIN.

# 9. WHAT IS THE POISSON DISTRIBUTION AND WHERE IS IT APPLIED?

  - THE POISSON DISTRIBUTION MODELS THE NUMBER OF EVENTS OCCURRING IN A FIXED INTERVAL OF TIME OR SPACE, GIVEN THAT THESE EVENTS HAPPEN INDEPENDENTLY AND AT A CONSTANT RATE. IT’S COMMONLY USED TO MODEL RARE EVENTS, SUCH AS THE NUMBER OF CAR ACCIDENTS AT A PARTICULAR INTERSECTION IN A DAY OR THE NUMBER OF EMAILS RECEIVED IN AN HOUR.

# 10. WHAT IS A CONTINUOUS UNIFORM DISTRIBUTION?

  - A CONTINUOUS UNIFORM DISTRIBUTION DESCRIBES A SITUATION WHERE ALL OUTCOMES IN A GIVEN RANGE ARE EQUALLY LIKELY. FOR EXAMPLE, IF YOU RANDOMLY PICK A NUMBER BETWEEN 0 AND 1, EVERY NUMBER IN THAT RANGE HAS AN EQUAL PROBABILITY OF BEING SELECTED. IT’S OFTEN VISUALIZED AS A RECTANGULAR SHAPE IN PROBABILITY PLOTS.

# 11. WHAT ARE THE CHARACTERISTICS OF A NORMAL DISTRIBUTION?

  • A NORMAL DISTRIBUTION IS A CONTINUOUS PROBABILITY DISTRIBUTION THAT IS SYMMETRIC ABOUT ITS MEAN. THE CHARACTERISTICS INCLUDE:

  • THE DISTRIBUTION IS BELL-SHAPED.

  • IT IS DEFINED BY TWO PARAMETERS: THE MEAN (Μ) AND THE STANDARD DEVIATION (Σ).

  • ABOUT 68% OF THE DATA FALLS WITHIN ONE STANDARD DEVIATION OF THE MEAN, 95% WITHIN TWO STANDARD DEVIATIONS, AND 99.7% WITHIN THREE STANDARD DEVIATIONS.

# 12. WHAT IS THE STANDARD NORMAL DISTRIBUTION, AND WHY IS IT IMPORTANT?

  -  THE STANDARD NORMAL DISTRIBUTION IS A SPECIAL CASE OF THE NORMAL DISTRIBUTION WHERE THE MEAN IS 0 AND THE STANDARD DEVIATION IS 1. IT IS IMPORTANT BECAUSE IT PROVIDES A COMMON REFERENCE FOR COMPARING DIFFERENT NORMAL DISTRIBUTIONS. VALUES IN A STANDARD NORMAL DISTRIBUTION ARE REFERRED TO AS Z-SCORES.

# 13. WHAT IS THE CENTRAL LIMIT THEOREM (CLT), AND WHY IS IT CRITICAL IN STATISTICS?

  -  THE CENTRAL LIMIT THEOREM STATES THAT, FOR A LARGE ENOUGH SAMPLE SIZE, THE SAMPLING DISTRIBUTION OF THE SAMPLE MEAN WILL BE APPROXIMATELY NORMALLY DISTRIBUTED, REGARDLESS OF THE ORIGINAL DISTRIBUTION OF THE DATA. THIS IS CRUCIAL BECAUSE IT ALLOWS US TO MAKE INFERENCES ABOUT A POPULATION FROM A SAMPLE, EVEN WHEN THE POPULATION DISTRIBUTION IS NOT NORMAL.

# 14. HOW DOES THE CENTRAL LIMIT THEOREM RELATE TO THE NORMAL DISTRIBUTION?

  - THE CLT IS CLOSELY RELATED TO THE NORMAL DISTRIBUTION BECAUSE IT TELLS US THAT THE DISTRIBUTION OF THE SAMPLE MEAN WILL APPROACH A NORMAL DISTRIBUTION AS THE SAMPLE SIZE INCREASES. THIS IS TRUE EVEN IF THE ORIGINAL DATA DISTRIBUTION IS NOT NORMAL, MAKING THE NORMAL DISTRIBUTION AN ESSENTIAL TOOL IN STATISTICAL INFERENCE.

# 15. WHAT IS THE APPLICATION OF Z STATISTICS IN HYPOTHESIS TESTING?

  **Z STATISTICS** (OR **Z-TEST**) IS USED IN HYPOTHESIS TESTING WHEN THE POPULATION STANDARD DEVIATION IS KNOWN AND THE SAMPLE SIZE IS LARGE (TYPICALLY $N \GEQ 30$). IT HELPS DETERMINE WHETHER TO REJECT THE NULL HYPOTHESIS.

#### APPLICATIONS INCLUDE:

  * **TESTING POPULATION MEANS**: WHEN COMPARING A SAMPLE MEAN TO A KNOWN POPULATION MEAN.
  * **COMPARING TWO MEANS**: WHEN COMPARING MEANS FROM TWO INDEPENDENT LARGE SAMPLES.
  * **PROPORTION TESTS**: FOR COMPARING SAMPLE PROPORTIONS TO POPULATION PROPORTIONS.
  * **ASSUMPTIONS**:

  * NORMAL DISTRIBUTION OF THE POPULATION (OR LARGE SAMPLE SIZE DUE TO CENTRAL LIMIT THEOREM)
  * KNOWN POPULATION STANDARD DEVIATION.

# 16. HOW DO YOU CALCULATE A Z-SCORE, AND WHAT DOES IT REPRESENT?

  A **Z-SCORE** MEASURES HOW MANY STANDARD DEVIATIONS A DATA POINT IS FROM THE MEAN.

#### **FORMULA**:

$$
Z = \FRAC{X - \MU}{\SIGMA}
$$

WHERE:

* $X$ = INDIVIDUAL DATA POINT
* $\MU$ = POPULATION MEAN
* $\SIGMA$ = POPULATION STANDARD DEVIATION

#### **INTERPRETATION**:

* **Z = 0** → DATA POINT IS EXACTLY AT THE MEAN
* **Z > 0** → DATA POINT IS ABOVE THE MEAN
* **Z < 0** → DATA POINT IS BELOW THE MEAN

IT HELPS IN COMPARING DATA POINTS FROM DIFFERENT DISTRIBUTIONS AND IN STANDARDIZING SCORES.

# 17. WHAT ARE POINT ESTIMATES AND INTERVAL ESTIMATES IN STATISTICS?

#### **POINT ESTIMATE**:

A SINGLE VALUE USED TO ESTIMATE A POPULATION PARAMETER.

* EXAMPLE: USING A SAMPLE MEAN ($\BAR{X}$) TO ESTIMATE THE POPULATION MEAN ($\MU$).

#### **INTERVAL ESTIMATE**:

A RANGE OF VALUES (WITH A CONFIDENCE LEVEL) WITHIN WHICH THE POPULATION PARAMETER IS EXPECTED TO LIE.

* EXAMPLE: A 95% CONFIDENCE INTERVAL FOR THE MEAN MIGHT BE (45.2, 49.8), SUGGESTING WE'RE 95% CONFIDENT THE TRUE MEAN LIES IN THIS RANGE.

#### KEY DIFFERENCE:

* **POINT ESTIMATE** GIVES NO MEASURE OF UNCERTAINTY.
* **INTERVAL ESTIMATE** ACCOUNTS FOR SAMPLING VARIABILITY AND PROVIDES A RANGE WITH CONFIDENCE LEVELS (LIKE 90%, 95%, 99%).

# 18. WHAT IS THE SIGNIFICANCE OF CONFIDENCE INTERVALS IN STATISTICAL ANALYSIS?

  -  CONFIDENCE INTERVALS PROVIDE A RANGE OF VALUES WITHIN WHICH WE ARE CONFIDENT THAT THE TRUE POPULATION PARAMETER LIES. IT HELPS QUANTIFY THE UNCERTAINTY IN OUR ESTIMATES AND PROVIDES MORE CONTEXT THAN A POINT ESTIMATE ALONE. A 95% CONFIDENCE INTERVAL, FOR EXAMPLE, MEANS WE ARE 95% CONFIDENT THAT THE TRUE VALUE LIES WITHIN THAT INTERVAL.

# 19. WHAT IS THE RELATIONSHIP BETWEEN A Z-SCORE AND A CONFIDENCE INTERVAL?

  - THE Z-SCORE PLAYS A KEY ROLE IN CONSTRUCTING CONFIDENCE INTERVALS. FOR A GIVEN CONFIDENCE LEVEL, THE Z-SCORE REPRESENTS THE NUMBER OF STANDARD DEVIATIONS AWAY FROM THE MEAN. IT’S USED IN THE FORMULA TO CALCULATE THE MARGIN OF ERROR FOR THE CONFIDENCE INTERVAL, WHICH GIVES THE RANGE AROUND THE SAMPLE ESTIMATE.

# 20. HOW ARE Z-SCORES USED TO COMPARE DIFFERENT DISTRIBUTIONS?

  - Z-SCORES STANDARDIZE DIFFERENT DATA SETS SO THEY CAN BE COMPARED ON THE SAME SCALE. BY CONVERTING DIFFERENT DISTRIBUTIONS TO Z-SCORES, WE CAN COMPARE THEIR RELATIVE POSITIONS AND SPREAD, EVEN IF THE ORIGINAL DISTRIBUTIONS HAVE DIFFERENT MEANS AND STANDARD DEVIATIONS.

# 21. WHAT ARE THE ASSUMPTIONS FOR APPLYING THE CENTRAL LIMIT THEOREM?

  • THE ASSUMPTIONS FOR APPLYING THE CENTRAL LIMIT THEOREM INCLUDE:

  • THE SAMPLE SIZE SHOULD BE LARGE ENOUGH (TYPICALLY N > 30).

  • THE SAMPLES SHOULD BE INDEPENDENT OF EACH OTHER.

  • THE DATA SHOULD BE DRAWN FROM THE SAME POPULATION.

# 22. WHAT IS THE CONCEPT OF EXPECTED VALUE IN A PROBABILITY DISTRIBUTION?

  - THE EXPECTED VALUE (OR MEAN) OF A PROBABILITY DISTRIBUTION IS THE LONG-TERM AVERAGE OF ALL POSSIBLE OUTCOMES, WEIGHTED BY THEIR PROBABILITIES. IT REPRESENTS THE CENTER OF THE DISTRIBUTION AND PROVIDES A MEASURE OF THE “CENTRAL TENDENCY” OF THE DISTRIBUTION.

# 23. HOW DOES A PROBABILITY DISTRIBUTION RELATE TO THE EXPECTED OUTCOME OF A RANDOM VARIABLE?

  A **PROBABILITY DISTRIBUTION** IS FUNDAMENTALLY TIED TO THE **EXPECTED OUTCOME** OF A RANDOM VARIABLE BECAUSE IT DESCRIBES ALL THE POSSIBLE VALUES THE VARIABLE CAN TAKE AND HOW LIKELY EACH VALUE IS.

### 🔹 **RELATIONSHIP EXPLAINED:**

THE **EXPECTED VALUE** (OR **MEAN**) OF A RANDOM VARIABLE IS ESSENTIALLY THE **WEIGHTED AVERAGE** OF ALL POSSIBLE OUTCOMES, **WEIGHTED BY THEIR PROBABILITIES** FROM THE DISTRIBUTION.

### 🔸 **FOR A DISCRETE RANDOM VARIABLE**:

IF A RANDOM VARIABLE $X$ CAN TAKE ON VALUES $X_1, X_2, ..., X_N$ WITH PROBABILITIES $P(X_1), P(X_2), ..., P(X_N)$, THEN THE **EXPECTED VALUE** $E(X)$ IS:

$$
E(X) = \SUM_{I=1}^{N} X_I \CDOT P(X_I)
$$

### 🔸 **FOR A CONTINUOUS RANDOM VARIABLE**:

WITH A PROBABILITY DENSITY FUNCTION $F(X)$, THE EXPECTED VALUE IS:

$$
E(X) = \INT_{-\INFTY}^{\INFTY} X \CDOT F(X) \, DX
$$

### ✅ **IN ESSENCE**:

* THE **PROBABILITY DISTRIBUTION** TELLS YOU HOW LIKELY EACH VALUE OF A RANDOM VARIABLE IS.
* THE **EXPECTED VALUE** SUMMARIZES THE "LONG-RUN AVERAGE" OUTCOME IF YOU REPEATED THE EXPERIMENT MANY TIMES.

### 📌 EXAMPLE (DISCRETE):

LET $X$ REPRESENT THE OUTCOME OF ROLLING A FAIR SIX-SIDED DIE. THE PROBABILITY DISTRIBUTION IS:

$$
P(X = X) = \FRAC{1}{6} \TEXT{ FOR } X = 1, 2, 3, 4, 5, 6
$$

THEN:

$$
E(X) = \SUM_{X=1}^{6} X \CDOT \FRAC{1}{6} = \FRAC{1 + 2 + 3 + 4 + 5 + 6}{6} = 3.5
$$

SO, **3.5** IS THE EXPECTED OUTCOME OVER MANY ROLLS, EVEN THOUGH IT'S NOT AN ACTUAL DIE FACE.

```python
# 2. GENERATE A DISCRETE UNIFORM DISTRIBUTION USING PYTHON AND PLOT THE PROBABILITY MASS FUNCTION (PMF).

IMPORT NUMPY AS NP
IMPORT MATPLOTLIB.PYPLOT AS PLT

# GENERATE 1000 RANDOM VALUES FROM A DISCRETE UNIFORM DISTRIBUTION
SAMPLES = NP.RANDOM.RANDINT(1, 7, SIZE=1000)

# PLOT THE PROBABILITY MASS FUNCTION (PMF)
VALUES, COUNTS = NP.UNIQUE(SAMPLES, RETURN_COUNTS=TRUE)
PMF = COUNTS / LEN(SAMPLES)

PLT.BAR(VALUES, PMF, WIDTH=0.5, COLOR='BLUE', ALPHA=0.7)
PLT.TITLE('DISCRETE UNIFORM DISTRIBUTION PMF')
PLT.XLABEL('VALUE')
PLT.YLABEL('PROBABILITY')
PLT.SHOW()
```

```python
# 3. WRITE A PYTHON FUNCTION TO CALCULATE THE PROBABILITY DISTRIBUTION FUNCTION (PDF) OF A BERNOULLI DISTRIBUTION.

IMPORT SCIPY.STATS AS STATS
IMPORT MATPLOTLIB.PYPLOT AS PLT

DEF BERNOULLI_PDF(P, SIZE=1000):
    # P IS THE PROBABILITY OF SUCCESS
    X = [0, 1]
    PDF = [STATS.BERNOULLI.PMF(K, P) FOR K IN X]

    RETURN X, PDF

# PARAMETERS
P = 0.7  # PROBABILITY OF SUCCESS

X, PDF = BERNOULLI_PDF(P)
PLT.BAR(X, PDF, COLOR='GREEN', ALPHA=0.6)
PLT.TITLE('BERNOULLI DISTRIBUTION PDF')
PLT.XLABEL('OUTCOME (0 OR 1)')
PLT.YLABEL('PROBABILITY')
PLT.SHOW()
```

```python
# 4. WRITE A PYTHON SCRIPT TO SIMULATE A BINOMIAL DISTRIBUTION WITH N=10 AND P=0.5, THEN PLOT ITS HISTOGRAM.

IMPORT NUMPY AS NP
IMPORT MATPLOTLIB.PYPLOT AS PLT

# PARAMETERS FOR BINOMIAL DISTRIBUTION
N = 10  # NUMBER OF TRIALS
P = 0.5  # PROBABILITY OF SUCCESS

# SIMULATE 1000 TRIALS OF THE BINOMIAL DISTRIBUTION
SAMPLES = NP.RANDOM.BINOMIAL(N, P, SIZE=1000)

# PLOT HISTOGRAM
PLT.HIST(SAMPLES, BINS=NP.ARANGE(-0.5, N+1.5, 1), DENSITY=TRUE, ALPHA=0.7, COLOR='ORANGE')
PLT.TITLE('BINOMIAL DISTRIBUTION (N=10, P=0.5)')
PLT.XLABEL('NUMBER OF SUCCESSES')
PLT.YLABEL('PROBABILITY')
PLT.SHOW()
```

```python
# 6.  WRITE A PYTHON PROGRAM TO CALCULATE AND PLOT THE CUMULATIVE DISTRIBUTION FUNCTION (CDF) OF A DISCRETE UNIFORM DISTRIBUTION.

IMPORT NUMPY AS NP
IMPORT MATPLOTLIB.PYPLOT AS PLT

# GENERATE 1000 RANDOM VALUES FROM A DISCRETE UNIFORM DISTRIBUTION
SAMPLES = NP.RANDOM.RANDINT(1, 7, SIZE=1000)

# CALCULATE CDF
VALUES, COUNTS = NP.UNIQUE(SAMPLES, RETURN_COUNTS=TRUE)
CDF = NP.CUMSUM(COUNTS) / LEN(SAMPLES)

# PLOT CDF
PLT.PLOT(VALUES, CDF, MARKER='O', COLOR='RED')
PLT.TITLE('DISCRETE UNIFORM DISTRIBUTION CDF')
PLT.XLABEL('VALUE')
PLT.YLABEL('CUMULATIVE PROBABILITY')
PLT.SHOW()
```

```python
# 9.  WRITE A PYTHON FUNCTION TO CALCULATE Z-SCORES FROM A DATASET AND PLOT THEM.

IMPORT NUMPY AS NP
IMPORT MATPLOTLIB.PYPLOT AS PLT

DEF CALCULATE_Z_SCORES(DATA):
    MEAN = NP.MEAN(DATA)
    STD = NP.STD(DATA)
    Z_SCORES = (DATA - MEAN) / STD
    RETURN Z_SCORES

# SIMULATE SOME DATA FROM A NORMAL DISTRIBUTION
DATA = NP.RANDOM.NORMAL(0, 1, 1000)

# CALCULATE Z-SCORES
Z_SCORES = CALCULATE_Z_SCORES(DATA)

# PLOT Z-SCORES
PLT.HIST(Z_SCORES, BINS=30, DENSITY=TRUE, ALPHA=0.7, COLOR='PINK')
PLT.TITLE('Z-SCORES')
PLT.XLABEL('Z-SCORE')
PLT.YLABEL('DENSITY')
PLT.SHOW()
```

```python
# 13. GENERATE RANDOM VARIABLES AND CALCULATE THEIR CORRESPONDING PROBABILITIES USING THE BINOMIAL DISTRIBUTION.

IMPORT NUMPY AS NP
IMPORT MATPLOTLIB.PYPLOT AS PLT
FROM SCIPY.STATS IMPORT BINOM

# PARAMETERS FOR BINOMIAL DISTRIBUTION
N = 10
P = 0.5

# GENERATE 1000 RANDOM SAMPLES FROM A BINOMIAL DISTRIBUTION
SAMPLES = NP.RANDOM.BINOMIAL(N, P, 1000)

# CALCULATE PROBABILITIES FOR EACH VALUE USING THE PMF
X = NP.ARANGE(0, N+1)
PMF = BINOM.PMF(X, N, P)

# PLOT HISTOGRAM OF SAMPLES AND THE PMF
PLT.HIST(SAMPLES, BINS=NP.ARANGE(-0.5, N+1.5, 1), DENSITY=TRUE, ALPHA=0.7, COLOR='GREEN')
PLT.PLOT(X, PMF, MARKER='O', COLOR='RED')
PLT.TITLE('BINOMIAL DISTRIBUTION PMF')
PLT.XLABEL('NUMBER OF SUCCESSES')
PLT.YLABEL('PROBABILITY')
PLT.SHOW()
```

```python
# 14. WRITE A PYTHON PROGRAM TO CALCULATE THE Z-SCORE FOR A GIVEN DATA POINT AND COMPARE IT TO A STANDARD NORMAL DISTRIBUTION.

IMPORT NUMPY AS NP
FROM SCIPY.STATS IMPORT NORM

# DATA
DATA_POINT = 2
MEAN = 0
STD = 1

# CALCULATE Z-SCORE
Z_SCORE = (DATA_POINT - MEAN) / STD
PRINT(F"Z-SCORE FOR DATA POINT {DATA_POINT}: {Z_SCORE}")

# COMPARE WITH STANDARD NORMAL DISTRIBUTION
PROBABILITY = NORM.CDF(Z_SCORE)
PRINT(F"PROBABILITY OF Z-SCORE {Z_SCORE} IN STANDARD NORMAL DISTRIBUTION: {PROBABILITY}")
```

```python
# 15. IMPLEMENT HYPOTHESIS TESTING USING Z-STATISTICS FOR A SAMPLE DATASET.

IMPORT NUMPY AS NP
FROM SCIPY.STATS IMPORT NORM

# SAMPLE DATA
SAMPLE_DATA = NP.RANDOM.NORMAL(50, 10, 100)

# PARAMETERS FOR POPULATION
POPULATION_MEAN = 50
POPULATION_STD = 10

# CALCULATE SAMPLE MEAN AND STANDARD ERROR
SAMPLE_MEAN = NP.MEAN(SAMPLE_DATA)
SAMPLE_STD = NP.STD(SAMPLE_DATA)
SAMPLE_SIZE = LEN(SAMPLE_DATA)

# Z-STATISTIC
Z_STAT = (SAMPLE_MEAN - POPULATION_MEAN) / (SAMPLE_STD / NP.SQRT(SAMPLE_SIZE))
PRINT(F"Z-STATISTIC: {Z_STAT}")

# P-VALUE FOR TWO-TAILED TEST
P_VALUE = 2 * (1 - NORM.CDF(ABS(Z_STAT)))
PRINT(F"P-VALUE: {P_VALUE}")
```
