# STATISTICS PART 2

### 1. WHAT IS HYPOTHESIS TESTING IN STATISTICS?
ANS:-HYPOTHESIS TESTING IS A METHOD USED TO DECIDE WHETHER THERE'S ENOUGH EVIDENCE TO SUPPORT A SPECIFIC CLAIM ABOUT A POPULATION. IT INVOLVES FORMULATING A NULL AND AN ALTERNATIVE HYPOTHESIS, THEN USING SAMPLE DATA TO ACCEPT OR REJECT THE NULL.

### 2. WHAT IS THE NULL HYPOTHESIS, AND HOW DOES IT DIFFER FROM THE ALTERNATIVE HYPOTHESIS?
ANS:-THE NULL HYPOTHESIS (H₀) STATES THERE IS NO EFFECT OR NO DIFFERENCE. THE ALTERNATIVE HYPOTHESIS (H₁ OR HA) IS WHAT YOU WANT TO PROVE — THAT THERE IS AN EFFECT OR A DIFFERENCE. HYPOTHESIS TESTING AIMS TO REJECT H₀ IN FAVOR OF H₁ IF EVIDENCE SUPPORTS IT.

### 3. WHAT IS THE SIGNIFICANCE LEVEL IN HYPOTHESIS TESTING, AND WHY IS IT IMPORTANT?
ANS:-THE SIGNIFICANCE LEVEL (Α) IS THE THRESHOLD FOR REJECTING THE NULL HYPOTHESIS, USUALLY SET AT 0.05. IT REPRESENTS THE PROBABILITY OF MAKING A TYPE I ERROR — REJECTING A TRUE NULL. IT CONTROLS HOW STRICT YOUR TEST IS.

### 4. WHAT DOES A P-VALUE REPRESENT IN HYPOTHESIS TESTING?
ANS:-THE P-VALUE IS THE PROBABILITY OF OBSERVING THE DATA (OR MORE EXTREME) ASSUMING THE NULL HYPOTHESIS IS TRUE. IT HELPS MEASURE HOW LIKELY YOUR SAMPLE RESULT IS UNDER H₀.

### 5. HOW DO YOU INTERPRET THE P-VALUE IN HYPOTHESIS TESTING?
ANS :-IF P ≤ Α: REJECT THE NULL HYPOTHESIS (EVIDENCE SUPPORTS H₁).

IF P > Α: FAIL TO REJECT THE NULL HYPOTHESIS (NOT ENOUGH EVIDENCE).
LOWER P-VALUES INDICATE STRONGER EVIDENCE AGAINST H₀.

###  6. WHAT ARE TYPE I AND TYPE II ERRORS IN HYPOTHESIS TESTING?
ANS:-TYPE I ERROR: REJECTING A TRUE NULL HYPOTHESIS (FALSE POSITIVE).

TYPE II ERROR: FAILING TO REJECT A FALSE NULL HYPOTHESIS (FALSE NEGATIVE).
Α CONTROLS TYPE I ERROR; Β RELATES TO TYPE II ERROR.

###  7. WHAT IS THE DIFFERENCE BETWEEN A ONE-TAILED AND A TWO-TAILED TEST IN HYPOTHESIS TESTING?
ANS:-ONE-TAILED TEST CHECKS FOR EFFECT IN ONE DIRECTION (E.G., GREATER THAN).

TWO-TAILED TEST CHECKS FOR EFFECT IN BOTH DIRECTIONS (E.G., NOT EQUAL).
CHOICE DEPENDS ON YOUR HYPOTHESIS.

### 8. WHAT IS THE Z-TEST, AND WHEN IS IT USED IN HYPOTHESIS TESTING?
ANS:-A Z-TEST IS USED TO COMPARE A SAMPLE TO A POPULATION WHEN THE POPULATION STANDARD DEVIATION IS KNOWN AND SAMPLE SIZE IS LARGE. IT HELPS TEST MEANS OR PROPORTIONS.

### 9. HOW DO YOU CALCULATE THE Z-SCORE, AND WHAT DOES IT REPRESENT IN HYPOTHESIS TESTING?
ANS:-Z = (X - Μ) / Σ.
IT MEASURES HOW FAR A SAMPLE VALUE IS FROM THE POPULATION MEAN IN TERMS OF STANDARD DEVIATIONS. IT HELPS DECIDE IF THE RESULT IS STATISTICALLY SIGNIFICANT.

###  10. WHAT IS THE T-DISTRIBUTION, AND WHEN SHOULD IT BE USED INSTEAD OF THE NORMAL DISTRIBUTION?
ANS:-THE T-DISTRIBUTION IS USED WHEN THE SAMPLE SIZE IS SMALL (N < 30) AND THE POPULATION STANDARD DEVIATION IS UNKNOWN. IT HAS HEAVIER TAILS THAN THE NORMAL DISTRIBUTION.

### 11. WHAT IS THE DIFFERENCE BETWEEN A Z-TEST AND A T-TEST?
ANS:-Z-TEST: KNOWN POPULATION Σ, LARGE N

T-TEST: UNKNOWN Σ, SMALL N
USE Z WHEN MORE INFORMATION IS KNOWN; T WHEN DEALING WITH SAMPLE ESTIMATES.



### 12. WHAT IS THE T-TEST, AND HOW IS IT USED IN HYPOTHESIS TESTING?
ANS:-A T-TEST COMPARES SAMPLE MEANS WHEN THE POPULATION STANDARD DEVIATION IS UNKNOWN. IT HELPS TEST IF THE SAMPLE DIFFERS SIGNIFICANTLY FROM THE POPULATION OR ANOTHER SAMPLE.

### 13. WHAT IS THE RELATIONSHIP BETWEEN Z-TEST AND T-TEST IN HYPOTHESIS TESTING?
ANS:-BOTH TEST MEANS, BUT THE Z-TEST IS USED WHEN POPULATION Σ IS KNOWN, WHILE THE T-TEST IS USED WHEN Σ IS UNKNOWN. AS SAMPLE SIZE INCREASES, THE T-DISTRIBUTION APPROACHES THE NORMAL DISTRIBUTION.

### 14. WHAT IS A CONFIDENCE INTERVAL, AND HOW IS IT USED TO INTERPRET STATISTICAL RESULTS?
ANS:-A CONFIDENCE INTERVAL GIVES A RANGE WHERE THE POPULATION PARAMETER LIKELY LIES, WITH A CERTAIN CONFIDENCE (E.G., 95%). IF IT INCLUDES THE NULL VALUE, RESULTS MAY NOT BE STATISTICALLY SIGNIFICANT.

### 15. WHAT IS THE MARGIN OF ERROR, AND HOW DOES IT AFFECT THE CONFIDENCE INTERVAL?
ANS:-THE MARGIN OF ERROR IS THE AMOUNT ADDED/SUBTRACTED FROM THE SAMPLE ESTIMATE TO CREATE THE CONFIDENCE INTERVAL. A LARGER MARGIN = WIDER INTERVAL = MORE UNCERTAINTY.

### 16. HOW IS BAYES' THEOREM USED IN STATISTICS, AND WHAT IS ITS SIGNIFICANCE?
ANS:-BAYES' THEOREM UPDATES THE PROBABILITY OF A HYPOTHESIS BASED ON NEW EVIDENCE. IT'S WIDELY USED IN DECISION-MAKING, MACHINE LEARNING, AND MEDICAL DIAGNOSIS.

### 17. WHAT IS THE CHI-SQUARE DISTRIBUTION, AND WHEN IS IT USED?
ANS:-THE CHI-SQUARE DISTRIBUTION IS USED FOR TESTING RELATIONSHIPS BETWEEN CATEGORICAL VARIABLES. IT’S COMMON IN GOODNESS-OF-FIT TESTS AND INDEPENDENCE TESTS IN CONTINGENCY TABLES.

###  18. WHAT IS THE CHI-SQUARE GOODNESS OF FIT TEST, AND HOW IS IT APPLIED?
ANS:-IT TESTS IF OBSERVED FREQUENCIES MATCH EXPECTED FREQUENCIES. IF THE TEST STATISTIC IS HIGH (AND P-VALUE LOW), WE REJECT THE NULL THAT THE DATA FITS THE EXPECTED DISTRIBUTION.

### 19. WHAT IS THE F-DISTRIBUTION, AND WHEN IS IT USED IN HYPOTHESIS TESTING?
ANS:-THE F-DISTRIBUTION IS USED TO COMPARE TWO VARIANCES OR IN ANOVA. IT HELPS DETERMINE IF THE VARIABILITY BETWEEN GROUP MEANS IS LARGER THAN WITHIN-GROUP VARIABILITY.



### 20. WHAT IS AN ANOVA TEST, AND WHAT ARE ITS ASSUMPTIONS?
ANS:-ANOVA (ANALYSIS OF VARIANCE) COMPARES MEANS OF 3 OR MORE GROUPS. ASSUMPTIONS: NORMALITY, EQUAL VARIANCES, AND INDEPENDENCE OF OBSERVATIONS.

###  21. WHAT ARE THE DIFFERENT TYPES OF ANOVA TESTS?
ANS:-ONE-WAY ANOVA: ONE INDEPENDENT VARIABLE.

TWO-WAY ANOVA: TWO INDEPENDENT VARIABLES.

REPEATED MEASURES ANOVA: SAME SUBJECTS OVER TIME.
EACH EXAMINES GROUP MEAN DIFFERENCES UNDER DIFFERENT CONDITIONS.

### 22. WHAT IS THE F-TEST, AND HOW DOES IT RELATE TO HYPOTHESIS TESTING?
ANS:-AN F-TEST COMPARES TWO VARIANCES OR GROUP MEANS (IN ANOVA). A LARGE F VALUE SUGGESTS THE GROUP MEANS DIFFER SIGNIFICANTLY, HELPING DECIDE IF TO REJECT H₀.

###  3.WRITE A PYTHON FUNCTION TO CALCULATE THE PROBABILITY DISTRIBUTION FUNCTION (PDF) OF A BERNOULLI DISTRIBUTION.


```python
FROM SCIPY.STATS IMPORT BERNOULLI

DEF BERNOULLI_PDF(P):
    """
    CALCULATE THE PDF (PMF) OF A BERNOULLI DISTRIBUTION FOR P.
    RETURNS PROBABILITIES FOR OUTCOMES 0 AND 1.
    """
    IF NOT (0 <= P <= 1):
        RAISE VALUEERROR("PROBABILITY P MUST BE BETWEEN 0 AND 1")
    
    PROB_0 = BERNOULLI.PMF(0, P)
    PROB_1 = BERNOULLI.PMF(1, P)
    
    RETURN {0: PROB_0, 1: PROB_1}

# EXAMPLE USAGE
PDF = BERNOULLI_PDF(0.6)
PRINT("BERNOULLI PDF:", PDF)
```

### 14.WRITE A PYTHON PROGRAM TO CALCULATE THE Z-SCORE FOR A GIVEN DATA POINT AND COMPARE IT TO A STANDARD NORMAL DISTRIBUTION.

```python
IMPORT NUMPY AS NP
IMPORT MATPLOTLIB.PYPLOT AS PLT
FROM SCIPY.STATS IMPORT NORM

DEF CALCULATE_Z_SCORE(X, MEAN, STD_DEV):
    Z = (X - MEAN) / STD_DEV
    RETURN Z

# EXAMPLE DATA
X = 75             # DATA POINT
MEAN = 70          # POPULATION MEAN
STD_DEV = 10       # POPULATION STANDARD DEVIATION

# CALCULATE Z-SCORE
Z_SCORE = CALCULATE_Z_SCORE(X, MEAN, STD_DEV)
PRINT(F"Z-SCORE FOR X = {X}: {Z_SCORE:.2F}")

# PLOT STANDARD NORMAL DISTRIBUTION
X_VALUES = NP.LINSPACE(-4, 4, 1000)
Y_VALUES = NORM.PDF(X_VALUES, 0, 1)

PLT.PLOT(X_VALUES, Y_VALUES, LABEL="STANDARD NORMAL PDF", COLOR="BLUE")

# MARK THE Z-SCORE ON THE CURVE
PLT.AXVLINE(Z_SCORE, COLOR='RED', LINESTYLE='--', LABEL=F'Z = {Z_SCORE:.2F}')
PLT.FILL_BETWEEN(X_VALUES, 0, Y_VALUES, WHERE=(X_VALUES <= Z_SCORE), COLOR='RED', ALPHA=0.3)

PLT.TITLE("Z-SCORE ON STANDARD NORMAL DISTRIBUTION")
PLT.XLABEL("Z-SCORE")
PLT.YLABEL("PROBABILITY DENSITY")
PLT.LEGEND()
PLT.GRID(TRUE)
PLT.SHOW()
```

### 15.IMPLEMENT HYPOTHESIS TESTING USING Z-STATISTICS FOR A SAMPLE DATASET.

```python
IMPORT NUMPY AS NP
FROM SCIPY.STATS IMPORT NORM

# SAMPLE DATA
SAMPLE = [72, 68, 75, 70, 74, 69, 71, 73, 70, 72]

# POPULATION PARAMETERS
POPULATION_MEAN = 70
POPULATION_STD = 3  # ASSUMED KNOWN FOR Z-TEST

# SAMPLE STATISTICS
SAMPLE_MEAN = NP.MEAN(SAMPLE)
SAMPLE_SIZE = LEN(SAMPLE)

# Z-STATISTIC CALCULATION
Z_SCORE = (SAMPLE_MEAN - POPULATION_MEAN) / (POPULATION_STD / NP.SQRT(SAMPLE_SIZE))

# TWO-TAILED P-VALUE
P_VALUE = 2 * (1 - NORM.CDF(ABS(Z_SCORE)))

# PRINT RESULTS
PRINT(F"SAMPLE MEAN = {SAMPLE_MEAN:.2F}")
PRINT(F"Z-SCORE = {Z_SCORE:.4F}")
PRINT(F"P-VALUE = {P_VALUE:.4F}")

# SIGNIFICANCE LEVEL
ALPHA = 0.05

# DECISION
IF P_VALUE < ALPHA:
    PRINT("REJECT THE NULL HYPOTHESIS (SIGNIFICANT DIFFERENCE).")
ELSE:
    PRINT("FAIL TO REJECT THE NULL HYPOTHESIS (NO SIGNIFICANT DIFFERENCE).")
```

### 16.CREATE A CONFIDENCE INTERVAL FOR A DATASET USING PYTHON AND INTERPRET THE RESULT.

```python
IMPORT NUMPY AS NP
FROM SCIPY.STATS IMPORT NORM

# SAMPLE DATASET
DATA = [72, 68, 75, 70, 74, 69, 71, 73, 70, 72]

# SAMPLE STATISTICS
SAMPLE_MEAN = NP.MEAN(DATA)
SAMPLE_SIZE = LEN(DATA)
SAMPLE_STD = NP.STD(DATA, DDOF=1)  # SAMPLE STANDARD DEVIATION

# CONFIDENCE LEVEL
CONFIDENCE = 0.95
Z_CRITICAL = NORM.PPF((1 + CONFIDENCE) / 2)

# MARGIN OF ERROR
MARGIN_OF_ERROR = Z_CRITICAL * (SAMPLE_STD / NP.SQRT(SAMPLE_SIZE))

# CONFIDENCE INTERVAL
LOWER_BOUND = SAMPLE_MEAN - MARGIN_OF_ERROR
UPPER_BOUND = SAMPLE_MEAN + MARGIN_OF_ERROR

# OUTPUT RESULTS
PRINT(F"SAMPLE MEAN: {SAMPLE_MEAN:.2F}")
PRINT(F"CONFIDENCE INTERVAL (95%): ({LOWER_BOUND:.2F}, {UPPER_BOUND:.2F})")
```

### 17.GENERATE DATA FROM A NORMAL DISTRIBUTION, THEN CALCULATE AND INTERPRET THE CONFIDENCE INTERVAL FOR ITS MEAN.

```python
IMPORT NUMPY AS NP
FROM SCIPY.STATS IMPORT NORM

# STEP 1: GENERATE DATA FROM A NORMAL DISTRIBUTION
NP.RANDOM.SEED(0)  # FOR REPRODUCIBILITY
DATA = NP.RANDOM.NORMAL(LOC=100, SCALE=15, SIZE=50)  # MEAN=100, STD=15, SAMPLE SIZE=50

# STEP 2: CALCULATE SAMPLE STATISTICS
SAMPLE_MEAN = NP.MEAN(DATA)
SAMPLE_STD = NP.STD(DATA, DDOF=1)  # USE DDOF=1 FOR SAMPLE STD DEVIATION
N = LEN(DATA)

# STEP 3: CALCULATE 95% CONFIDENCE INTERVAL
CONFIDENCE = 0.95
Z_CRITICAL = NORM.PPF((1 + CONFIDENCE) / 2)  # Z ≈ 1.96 FOR 95%

MARGIN_OF_ERROR = Z_CRITICAL * (SAMPLE_STD / NP.SQRT(N))
CI_LOWER = SAMPLE_MEAN - MARGIN_OF_ERROR
CI_UPPER = SAMPLE_MEAN + MARGIN_OF_ERROR

# STEP 4: DISPLAY THE RESULT
PRINT(F"SAMPLE MEAN: {SAMPLE_MEAN:.2F}")
PRINT(F"95% CONFIDENCE INTERVAL: ({CI_LOWER:.2F}, {CI_UPPER:.2F})")
```

### 4.PERFORM A TWO-TAILED Z-TEST USING PYTHON AND VISUALIZE THE DECISION REGION ON A PLOT.

```python
IMPORT NUMPY AS NP
IMPORT MATPLOTLIB.PYPLOT AS PLT
FROM SCIPY.STATS IMPORT NORM

# --- PARAMETERS ---
SAMPLE = NP.RANDOM.NORMAL(LOC=102, SCALE=10, SIZE=30)  # SIMULATED SAMPLE
POPULATION_MEAN = 100
POPULATION_STD = 10  # KNOWN
ALPHA = 0.05

# --- Z-TEST ---
SAMPLE_MEAN = NP.MEAN(SAMPLE)
N = LEN(SAMPLE)
Z_SCORE = (SAMPLE_MEAN - POPULATION_MEAN) / (POPULATION_STD / NP.SQRT(N))
P_VALUE = 2 * (1 - NORM.CDF(ABS(Z_SCORE)))

# --- CRITICAL Z FOR TWO-TAILED TEST ---
Z_CRITICAL = NORM.PPF(1 - ALPHA/2)

# --- OUTPUT RESULTS ---
PRINT(F"SAMPLE MEAN: {SAMPLE_MEAN:.2F}")
PRINT(F"Z-SCORE: {Z_SCORE:.4F}")
PRINT(F"P-VALUE: {P_VALUE:.4F}")
PRINT(F"Z-CRITICAL (±): ±{Z_CRITICAL:.2F}")

# --- PLOT STANDARD NORMAL DISTRIBUTION ---
X = NP.LINSPACE(-4, 4, 1000)
Y = NORM.PDF(X)

PLT.FIGURE(FIGSIZE=(10, 5))
PLT.PLOT(X, Y, LABEL='STANDARD NORMAL CURVE', COLOR='BLUE')

# SHADE CRITICAL REGIONS
PLT.FILL_BETWEEN(X, Y, WHERE=(X <= -Z_CRITICAL), COLOR='RED', ALPHA=0.3, LABEL='REJECTION REGION (LEFT)')
PLT.FILL_BETWEEN(X, Y, WHERE=(X >= Z_CRITICAL), COLOR='RED', ALPHA=0.3, LABEL='REJECTION REGION (RIGHT)')

# MARK Z-SCORE
PLT.AXVLINE(Z_SCORE, COLOR='GREEN', LINESTYLE='--', LABEL=F'Z = {Z_SCORE:.2F}')

PLT.TITLE("TWO-TAILED Z-TEST VISUALIZATION")
PLT.XLABEL("Z-SCORE")
PLT.YLABEL("PROBABILITY DENSITY")
PLT.LEGEND()
PLT.GRID(TRUE)
PLT.SHOW()

# --- CONCLUSION ---
IF ABS(Z_SCORE) > Z_CRITICAL:
    PRINT("REJECT THE NULL HYPOTHESIS: SIGNIFICANT DIFFERENCE.")
ELSE:
    PRINT("FAIL TO REJECT THE NULL HYPOTHESIS: NO SIGNIFICANT DIFFERENCE.")
```

### 5. CREATE A PYTHON FUNCTION THAT CALCULATES AND VISUALIZES TYPE 1 AND TYPE 2 ERRORS DURING HYPOTHESIS TESTING.


```python
IMPORT NUMPY AS NP
IMPORT MATPLOTLIB.PYPLOT AS PLT
FROM SCIPY.STATS IMPORT NORM

DEF VISUALIZE_TYPE1_TYPE2_ERRORS(MU_NULL=100, MU_ALT=105, SIGMA=10, N=30, ALPHA=0.05):
    # STANDARD ERROR
    SE = SIGMA / NP.SQRT(N)

    # CRITICAL VALUE (TWO-TAILED)
    Z_CRIT = NORM.PPF(1 - ALPHA/2)

    # CRITICAL REGION BOUNDS UNDER H0
    CRIT_LOW = MU_NULL - Z_CRIT * SE
    CRIT_HIGH = MU_NULL + Z_CRIT * SE

    # X-AXIS RANGE COVERING BOTH DISTRIBUTIONS
    X = NP.LINSPACE(MU_NULL - 4*SE, MU_ALT + 4*SE, 1000)

    # PDF FOR H0 AND H1
    H0_PDF = NORM.PDF(X, MU_NULL, SE)
    H1_PDF = NORM.PDF(X, MU_ALT, SE)

    # PLOT BOTH DISTRIBUTIONS
    PLT.FIGURE(FIGSIZE=(10, 5))
    PLT.PLOT(X, H0_PDF, LABEL="H₀ DISTRIBUTION", COLOR='BLUE')
    PLT.PLOT(X, H1_PDF, LABEL="H₁ DISTRIBUTION", COLOR='GREEN')

    # TYPE I ERROR REGION (Α)
    PLT.FILL_BETWEEN(X, H0_PDF, WHERE=((X < CRIT_LOW) | (X > CRIT_HIGH)), COLOR='RED', ALPHA=0.3, LABEL='TYPE I ERROR (Α)')

    # TYPE II ERROR REGION (Β)
    PLT.FILL_BETWEEN(X, H1_PDF, WHERE=((X >= CRIT_LOW) & (X <= CRIT_HIGH)), COLOR='ORANGE', ALPHA=0.4, LABEL='TYPE II ERROR (Β)')

    # LABELS AND LEGEND
    PLT.AXVLINE(CRIT_LOW, COLOR='BLACK', LINESTYLE='--', LABEL=F"CRITICAL VALUE (LOW) = {CRIT_LOW:.2F}")
    PLT.AXVLINE(CRIT_HIGH, COLOR='BLACK', LINESTYLE='--', LABEL=F"CRITICAL VALUE (HIGH) = {CRIT_HIGH:.2F}")
    PLT.TITLE("TYPE I AND TYPE II ERRORS VISUALIZATION")
    PLT.XLABEL("SAMPLE MEAN")
    PLT.YLABEL("PROBABILITY DENSITY")
    PLT.LEGEND()
    PLT.GRID(TRUE)
    PLT.SHOW()

    # CALCULATE TYPE II ERROR (Β)
    BETA = NORM.CDF(CRIT_HIGH, LOC=MU_ALT, SCALE=SE) - NORM.CDF(CRIT_LOW, LOC=MU_ALT, SCALE=SE)
    POWER = 1 - BETA

    # OUTPUT SUMMARY
    PRINT(F"CRITICAL REGION: ({CRIT_LOW:.2F}, {CRIT_HIGH:.2F})")
    PRINT(F"TYPE I ERROR (Α): {ALPHA:.2F}")
    PRINT(F"TYPE II ERROR (Β): {BETA:.4F}")
    PRINT(F"POWER OF THE TEST (1 - Β): {POWER:.4F}")

# EXAMPLE USAGE
VISUALIZE_TYPE1_TYPE2_ERRORS()
```

### 9.WRITE A PYTHON FUNCTION TO CALCULATE THE CONFIDENCE INTERVAL FOR A SAMPLE MEAN AND EXPLAIN ITS SIGNIFICANCE.

```python
IMPORT NUMPY AS NP
FROM SCIPY.STATS IMPORT T

DEF CONFIDENCE_INTERVAL(DATA, CONFIDENCE=0.95):
    N = LEN(DATA)
    MEAN = NP.MEAN(DATA)
    STD_ERR = NP.STD(DATA, DDOF=1) / NP.SQRT(N)  # STANDARD ERROR OF THE MEAN

    # T CRITICAL VALUE FOR GIVEN CONFIDENCE LEVEL
    T_CRIT = T.PPF((1 + CONFIDENCE) / 2, DF=N-1)

    # MARGIN OF ERROR
    MARGIN = T_CRIT * STD_ERR

    LOWER_BOUND = MEAN - MARGIN
    UPPER_BOUND = MEAN + MARGIN

    PRINT(F"SAMPLE MEAN: {MEAN:.2F}")
    PRINT(F"{INT(CONFIDENCE*100)}% CONFIDENCE INTERVAL: ({LOWER_BOUND:.2F}, {UPPER_BOUND:.2F})")
    RETURN (LOWER_BOUND, UPPER_BOUND)

# EXAMPLE USAGE
NP.RANDOM.SEED(0)
SAMPLE_DATA = NP.RANDOM.NORMAL(LOC=100, SCALE=15, SIZE=30)
CONFIDENCE_INTERVAL(SAMPLE_DATA)
```

### 11.IMPLEMENT A BAYESIAN INFERENCE METHOD USING BAYES' THEOREM IN PYTHON AND EXPLAIN THE PROCESS.

```python
DEF BAYES_THEOREM(PRIOR, SENSITIVITY, FALSE_POSITIVE):
    """
    CALCULATE P(DISEASE | POSITIVE) USING BAYES' THEOREM.
    """
    # COMPLEMENTARY VALUES
    NOT_PRIOR = 1 - PRIOR
    SPECIFICITY = 1 - FALSE_POSITIVE

    # MARGINAL PROBABILITY OF TESTING POSITIVE
    P_POSITIVE = (SENSITIVITY * PRIOR) + (FALSE_POSITIVE * NOT_PRIOR)

    # POSTERIOR PROBABILITY
    POSTERIOR = (SENSITIVITY * PRIOR) / P_POSITIVE

    PRINT(F"PRIOR P(DISEASE): {PRIOR}")
    PRINT(F"SENSITIVITY P(POSITIVE | DISEASE): {SENSITIVITY}")
    PRINT(F"FALSE POSITIVE RATE P(POSITIVE | NO DISEASE): {FALSE_POSITIVE}")
    PRINT(F"POSTERIOR P(DISEASE | POSITIVE): {POSTERIOR:.4F}")

    RETURN POSTERIOR

# EXAMPLE USAGE
BAYES_THEOREM(PRIOR=0.01, SENSITIVITY=0.99, FALSE_POSITIVE=0.05)
```

## EXPLANATION OF THE PROCESS:
>PRIOR BELIEF: THERE'S A 1% CHANCE A PERSON HAS THE DISEASE.

>DATA (TEST RESULT): THE PERSON TESTS POSITIVE.

>LIKELIHOOD: THE TEST IS 99% ACCURATE FOR THOSE WHO HAVE THE DISEASE.

>FALSE POSITIVE: 5% OF HEALTHY PEOPLE TEST POSITIVE.

>POSTERIOR: AFTER SEEING THE TEST RESULT, THE PROBABILITY OF HAVING THE DISEASE RISES TO ~16.7% — NOT 99%, WHICH SURPRISES MANY!

### 12. PERFORM A CHI-SQUARE TEST FOR INDEPENDENCE BETWEEN TWO CATEGORICAL VARIABLES IN PYTHON.

```python
IMPORT PANDAS AS PD
FROM SCIPY.STATS IMPORT CHI2_CONTINGENCY

# STEP 1: CREATE A CONTINGENCY TABLE
# EXAMPLE: GENDER VS PREFERENCE
DATA = PD.DATAFRAME({
    'TEA': [30, 10],
    'COFFEE': [20, 40]
}, INDEX=['MALE', 'FEMALE'])

PRINT("CONTINGENCY TABLE:")
PRINT(DATA)

# STEP 2: PERFORM CHI-SQUARE TEST
CHI2, P, DOF, EXPECTED = CHI2_CONTINGENCY(DATA)

# STEP 3: OUTPUT RESULTS
PRINT("\NCHI-SQUARE TEST RESULTS:")
PRINT(F"CHI2 STATISTIC: {CHI2:.4F}")
PRINT(F"DEGREES OF FREEDOM: {DOF}")
PRINT(F"P-VALUE: {P:.4F}")
PRINT("\NEXPECTED FREQUENCIES:")
PRINT(PD.DATAFRAME(EXPECTED, INDEX=DATA.INDEX, COLUMNS=DATA.COLUMNS))

# STEP 4: INTERPRETATION
ALPHA = 0.05
IF P < ALPHA:
    PRINT("\NCONCLUSION: REJECT THE NULL HYPOTHESIS — VARIABLES ARE DEPENDENT.")
ELSE:
    PRINT("\NCONCLUSION: FAIL TO REJECT THE NULL HYPOTHESIS — VARIABLES ARE INDEPENDENT.")
```

### 13.WRITE A PYTHON PROGRAM TO CALCULATE THE EXPECTED FREQUENCIES FOR A CHI-SQUARE TEST BASED ON OBSERVED DATA.

```python
IMPORT PANDAS AS PD
IMPORT NUMPY AS NP

# STEP 1: DEFINE THE OBSERVED DATA (CONTINGENCY TABLE)
# EXAMPLE: PRODUCT PREFERENCE BY REGION
OBSERVED = PD.DATAFRAME({
    'PRODUCT A': [20, 30, 25],
    'PRODUCT B': [25, 15, 35]
}, INDEX=['REGION 1', 'REGION 2', 'REGION 3'])

PRINT("OBSERVED FREQUENCIES:")
PRINT(OBSERVED)

# STEP 2: CALCULATE ROW TOTALS, COLUMN TOTALS, AND GRAND TOTAL
ROW_TOTALS = OBSERVED.SUM(AXIS=1)
COL_TOTALS = OBSERVED.SUM(AXIS=0)
GRAND_TOTAL = OBSERVED.VALUES.SUM()

# STEP 3: CALCULATE EXPECTED FREQUENCIES
EXPECTED = PD.DATAFRAME(INDEX=OBSERVED.INDEX, COLUMNS=OBSERVED.COLUMNS)

FOR ROW IN OBSERVED.INDEX:
    FOR COL IN OBSERVED.COLUMNS:
        EXPECTED.LOC[ROW, COL] = (ROW_TOTALS[ROW] * COL_TOTALS[COL]) / GRAND_TOTAL

EXPECTED = EXPECTED.ASTYPE(FLOAT)

# STEP 4: DISPLAY RESULTS
PRINT("\NEXPECTED FREQUENCIES:")
PRINT(EXPECTED.ROUND(2))
```

### 14. PERFORM A GOODNESS-OF-FIT TEST USING PYTHON TO COMPARE THE OBSERVED DATA TO AN EXPECTED DISTRIBUTION.


```python
IMPORT NUMPY AS NP
FROM SCIPY.STATS IMPORT CHISQUARE

# STEP 1: OBSERVED FREQUENCIES (E.G., OUTCOMES OF ROLLING A DIE)
OBSERVED = NP.ARRAY([18, 20, 17, 22, 19, 24])  # OBSERVED COUNTS FOR EACH FACE OF A DIE

# STEP 2: EXPECTED FREQUENCIES (UNIFORM DISTRIBUTION: EQUAL CHANCE FOR EACH FACE)
EXPECTED = NP.ARRAY([20, 20, 20, 20, 20, 20])

# STEP 3: PERFORM CHI-SQUARE GOODNESS-OF-FIT TEST
CHI_STAT, P_VALUE = CHISQUARE(F_OBS=OBSERVED, F_EXP=EXPECTED)

# STEP 4: PRINT RESULTS
PRINT("OBSERVED FREQUENCIES:", OBSERVED)
PRINT("EXPECTED FREQUENCIES:", EXPECTED)
PRINT(F"CHI-SQUARE STATISTIC: {CHI_STAT:.4F}")
PRINT(F"P-VALUE: {P_VALUE:.4F}")

# STEP 5: INTERPRET RESULTS
ALPHA = 0.05
IF P_VALUE < ALPHA:
    PRINT("REJECT THE NULL HYPOTHESIS: OBSERVED DATA DOES NOT FIT THE EXPECTED DISTRIBUTION.")
ELSE:
    PRINT("FAIL TO REJECT THE NULL HYPOTHESIS: OBSERVED DATA FITS THE EXPECTED DISTRIBUTION.")
```
